# Color Selection Code for Lane Lines Detection

---

### Importing Libraries

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

### Read in the sample image and print some stats

```python
image = mpimg.imread('1-Test-Image-for-Lane-Lines.jpg')
print('This image is: ',type(image), 'with dimensions:', image.shape)
```

### The original image

```python
print('Original Image')
plt.imshow(image)
plt.show()
```

### Grab the x and y size

```python
ysize = image.shape[0]
xsize = image.shape[1]
```

### Copy image

```python
color_select = np.copy(image)
```

### Color thresholds to filter

```python
red_threshold = 200
green_threshold = 200
blue_threshold = 200
rgb_threshold = [red_threshold, green_threshold, blue_threshold]
```

### Identify pixels below the threshold

```python
thresholds = (image[:,:,0] < rgb_threshold[0]) \
            | (image[:,:,1] < rgb_threshold[1]) \
            | (image[:,:,2] < rgb_threshold[2])
color_select[thresholds] = [0,0,0]
```

### Show the result image after color selection

```python
print('Result Image after color selection')
plt.imshow(color_select)
plt.show()
```

### Save the result image

```python
mpimg.imsave("1-Color-Selection-Code-Result.png", color_select)
```
